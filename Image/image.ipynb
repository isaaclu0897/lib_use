```python
from skimage import data
import matplotlib.pyplot as plt

image = data.astronaut()

plt.imshow(image)
plt.show()
```

```python
plt.figure(num='baby', figsize=(10, 18), dpi=100)

plt.subplot(521)
plt.title('1')
plt.imshow(image[:,:,2], plt.cm.Blues_r)
plt.colorbar()

plt.subplot(522)
plt.title('2')
plt.imshow(image[:,:,0],plt.cm.jet)
plt.axis('off')
plt.colorbar()


plt.subplot(523)
plt.title('3')
plt.imshow(image[:,:,1],plt.cm.BrBG)
plt.axis('off')
plt.colorbar()


plt.subplot(524)
plt.title('4')
plt.imshow(image[:,:,2],plt.cm.autumn)
plt.axis('off')
plt.colorbar()


plt.show()

```
